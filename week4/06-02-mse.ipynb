# Mean Square Error (MSE) Loss

```python
import torch
```

```python
# MSE 함수 정의
def mse(x_hat, x):
    # x_hat, x shape: (batch_size, dim)
    y = ((x - x_hat)**2).mean()
    return y
```

```python
# 테스트 데이터
x = torch.FloatTensor([[1, 1],
                       [2, 2]])
x_hat = torch.FloatTensor([[0, 0],
                           [0, 0]])

print(x.size(), x_hat.size())
```

```python
# MSE 계산
mse(x_hat, x)
```

```python
import torch.nn.functional as F
```

```python
# PyTorch 내장 MSE
F.mse_loss(x_hat, x)
```
