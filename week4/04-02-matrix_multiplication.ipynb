## Matrix Multiplication

```python
import torch  # line
```

```python
x = torch.FloatTensor([[1, 2],  # line
                       [3, 4],  # line
                       [5, 6]])  # line
y = torch.FloatTensor([[1, 2],  # line
                       [1, 2]])  # line

print(x.size(), y.size())  # line
```

```python
z = torch.matmul(x, y)  # line
print(z.size())  # line
```

```python
x = torch.FloatTensor([[[1, 2],  # line
                        [3, 4],  # line
                        [5, 6]],  # line
                       [[7, 8],  # line
                        [9, 10],  # line
                        [11, 12]],  # line
                       [[13, 14],  # line
                        [15, 16],  # line
                        [17, 18]]])  # line
y = torch.FloatTensor([[[1, 2, 2],  # line
                        [1, 2, 2]],  # line
                       [[1, 3, 3],  # line
                        [1, 3, 3]],  # line
                       [[1, 4, 4],  # line
                        [1, 4, 4]]])  # line

print(x.size(), y.size())  # line
```

```python
z = torch.bmm(x, y)  # line
print(z.size())  # line
```
