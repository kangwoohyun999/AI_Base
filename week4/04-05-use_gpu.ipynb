# How to use GPU

```python
import torch
import torch.nn as nn
```

```python
# CUDA tensor 생성
x = torch.cuda.FloatTensor(2, 2)
x
```

```python
# CPU tensor를 GPU로 이동
x = torch.FloatTensor(2, 2)
x = x.cuda()
x
```

```python
# device 지정
d = torch.device('cuda:0')
x = x.to(device=d)
x
```

```python
# 모델을 GPU로 이동
linear = nn.Linear(2, 2)
linear = linear.cuda()
```
