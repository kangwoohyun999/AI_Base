# PyTorch Tensor Operations

```python
import torch  # PyTorch 불러오기
```

## Arithmetic Element-wise Operations

```python
a = torch.FloatTensor([[1, 2],      # 텐서 a 생성
                       [3, 4]])
b = torch.FloatTensor([[2, 2],      # 텐서 b 생성
                       [3, 3]])
```

```python
a + b       # 덧셈
```

```python
a - b       # 뺄셈
```

```python
a * b       # 원소별 곱셈
```

```python
a / b       # 나눗셈
```

```python
a == b      # 비교 (같은지)
```

```python
a != b      # 비교 (다른지)
```

```python
a ** b      # 거듭제곱
```

## Inplace Operations

```python
print(a)        # 원본 출력
print(a.mul(b))     # 곱셈 (새 텐서 반환)
print(a)        # 원본 유지
print(a.mul_(b))         # in-place 곱셈 (a 변경)
print(a)         # 변경된 값 출력
```

## Sum, Mean (Dimension Reducing Operations)

```python
x = torch.FloatTensor([[1, 2],      # 텐서 생성
                       [3, 4]])
```

```python
print(x.sum())      # 전체 합
print(x.mean())      # 평균
```

```python
print(x.sum(dim=0))      # 열 기준 합
print(x.sum(dim=-1))    # 행 기준 합
```

## Broadcast in Operations

What we did before,

```python
x = torch.FloatTensor([[1, 2]])     # 1x2 텐서
y = torch.FloatTensor([[4, 8]])      # 1x2 텐서

print(x.size())     # 크기 출력
print(y.size())     # 크기 출력
```

```python
z = x + y       # 같은 shape 연산
print(z)        #원본 출력
print(z.size())     # 크기 출력
```

Broadcast feature provides operations between different shape of tensors.

### Tensor + Scalar

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]])
y = 1       # 스칼라

print(x.size())
```

```python
z = x + y       # broadcasting
print(z)
print(z.size())
```

### Tensor + Vector

```python
x = torch.FloatTensor([[1, 2],
                       [4, 8]])
y = torch.FloatTensor([3,       # 1차원 텐서
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y       # broadcasting
print(z)
print(z.size())
```

```python
x = torch.FloatTensor([[[1, 2]]])       # 3D 텐서
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y       # broadcasting
print(z)
print(z.size())
```

### Tensor + Tensor

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[3],
                       [5]])

print(x.size())
print(y.size())
```

```python
z = x + y       # broadcasting
print(z)
print(z.size())
```

Note that you need to be careful before using broadcast feature.

### Failure Case

```python
x = torch.FloatTensor([[[1, 2],
                        [4, 8]]])
y = torch.FloatTensor([[1, 2, 3],
                       [4, 5, 6],
                       [7, 8, 9]])

print(x.size())
print(y.size())

z = x + y       # broadcasting (불가능한 경우 에러 발생 가능)
```
