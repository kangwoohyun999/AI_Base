# Hello PyTorch!

Import PyTorch.

```python
import torch# PyTorch 라이브러리 불러오기
```

Define PyTorch float tensors.

```python
a = torch.FloatTensor([[1, 2],      # float 타입 텐서 생성
                       [3, 4]])
b = torch.FloatTensor([[1, 2],      # float 타입 텐서 생성
                       [1, 2]])
```

Matrix multiplication example.

```python
c = torch.matmul(a, b) #행렬 곱
c
```
